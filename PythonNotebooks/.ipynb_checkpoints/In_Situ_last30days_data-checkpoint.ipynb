<h2>PLOT THE INFORMATION AVAILABLE ON THE INDEX FILE</h2>

```python
#connect to CMEMS FTP
with ftputil.FTPHost(host, user, password) as ftp_host: 
    #open the index file to read
    with ftp_host.open("Core"+'/'+product_name+'/'+index_file, "r") as indexfile:
        pie_raw = pd.read_csv(indexfile, skiprows=5) #load it as pandas dataframe
```

<H2>DERIVED INFORMATION FROM FILENAME</H2>

```python
def filename_convention(filename):
    elements = filename.split('.nc')[0].split('_')
    if len(elements[1]) == 2: #history file RR_XX_YY_CODE<_ZZZ>.nc
        if len(elements) == 4:
            return ['region', 'data', 'platform', 'code', 'subsettingcode']
        else:
            return ['region', 'data', 'platform', 'code']
    elif elements[1] == 'LATEST': #RR_LATEST_XX_YY_CODE_YYYYMMDD.nc
        return ['region', 'fixedname', 'data', 'platform', 'code', 'timestamp']
    else: #monthly file RR_YYYYMM_XX_YY_CODE.nc
        return ['region', 'timestamp', 'data', 'platform', 'code']
```

```python
key = len(pie_raw['file_name'][0].split('/'))-1
df = pie_raw['file_name'].str.split('/',expand=True)
df[['filename','extension']] = df[key].str.split('.',expand=True)
df.head()
```

```python
headers = filename_convention(pie_raw['file_name'][0])
df[headers] = df['filename'].str.split('_',expand=True)
df.head()
```

<H3>REGIONAL PRODUCTION UNIT BREAKDOWN</H3>
