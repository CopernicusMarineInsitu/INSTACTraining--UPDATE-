<H2>MADATORY PYTHON LIBRARIES</H2>

```python
import ftputil
import numpy as np
import pandas as pd
from shapely.geometry import box
import folium
from folium import plugins
import datetime
```

<h2> SET YOUR CREDENTIALS</h2>

```python
user = '' #type CMEMS user name
password = '' #type CMEMS password
```

<h2> TARGET A PRODUCT, HOST AND INDEX FILE</h2>

```python
product_name = 'INSITU_MED_NRT_OBSERVATIONS_013_035' #type aimed In Situ product 
host = 'nrt.cmems-du.eu' #type aimed host (nrt.cmems-du.eu or my.cmems-du)
index_file = 'index_latest.txt' #type aimed index file 
```

<h2>INDEX FILE RAW INFORMATION</h2>

```python
#connect to CMEMS FTP
with ftputil.FTPHost(host, user, password) as ftp_host: 
    #open the index file to read
    with ftp_host.open("Core"+'/'+product_name+'/'+index_file, "r") as indexfile:
        raw_index_info = pd.read_csv(indexfile, skiprows=5) #load it as pandas dataframe
raw_index_info.head()
```

<h3>HEATMAP</h3>

```python
lats, lons, mags = [], [], []
for geospatial_lat_min, geospatial_lat_max, geospatial_lon_min, geospatial_lon_max, ini, end in zip(raw_index_info['geospatial_lat_min'].tolist(), raw_index_info['geospatial_lat_max'].tolist(), raw_index_info['geospatial_lon_min'].tolist(), raw_index_info['geospatial_lon_max'].tolist(), raw_index_info['time_coverage_start'].tolist(), raw_index_info['time_coverage_end'].tolist()):
    bounding_box = box(geospatial_lon_min, geospatial_lat_min, geospatial_lon_max, geospatial_lat_max)
    lats.append(bounding_box.centroid.y)
    lons.append(bounding_box.centroid.x)
    date_format = "%Y-%m-%dT%H:%M:%SZ" 
    ini = datetime.datetime.strptime(ini, date_format)
    end = datetime.datetime.strptime(end, date_format)     
    mags.append((end-ini).seconds)
    
map = folium.Map(location=[np.mean(lats), np.mean(lons)], zoom_start=5)
map.add_child(plugins.HeatMap(zip(lats, lons, mags), radius = 10))
```
